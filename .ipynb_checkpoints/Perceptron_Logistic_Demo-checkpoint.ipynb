This is worked example of a perceptron with a binary step function as well as with a sigmoid activation (logistic function) similar to a logistic regression problem.

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

A useful tip for reproducibility, i.e. to be able to reproduce our results similarily every time, is to set the seed for the random number generator. 

```python
np.random.seed(200)
```

Let's generate some data. We are going two models which approximate the binary `&` (and) and `^` (xor) operators. The and operator is linearly seperable, but the xor operator is not, so we expect the perceptron to work for and, but struggle with xor.

```python
n = 100
```

```python
data = (np.random.rand(n, 2) > 0.5)
t_and = (data[:, 0] & data[:, 1]).astype(int)
t_xor = (data[:, 0] ^ data[:, 1]).astype(int)
data = data.astype(float)
data += np.random.randn(n, 2) * 0.3
X = np.hstack([np.ones((n, 1)), data])
```

Let's have a look at the data.

```python
plt.scatter(data[t_and == 0, 0], data[t_and == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_and == 1, 0], data[t_and == 1, 1], color="red", edgecolors='black');
plt.xlabel('$x$');
plt.ylabel('$y$');
plt.title('Data: Noisy Logical AND-Operator');
```

The orange labelled points are located more or less in the top right corner, where $x = 1, y = 1$. This corresponds to the implemented logical `and` operator. The noise is added to make the data more interesting. In a real world scenario, we often encounter measurement errors or unexplained variation in our data. Adding noise makes simulates this uncertainty.

```python
plt.scatter(data[t_xor == 0, 0], data[t_xor == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_xor == 1, 0], data[t_xor == 1, 1], color="red", edgecolors='black');
plt.xlabel('$x$');
plt.ylabel('$y$');
plt.title('Data: Noisy Logical XOR-Operator');
```

The 'xor' operator is more tricky, as the two classes are not linearly seperable. 

Now, we implement a loss, or cost function. The loss function is the absolute difference between the target and the predictions, i.e.

$$
C(t, \hat t) = \lvert t - \hat t\rvert
$$

Target values are given by $t$, and approximations are given by $\hat t$. Sometimes the loss function is denoted as $L$, other times it's denoted as $C$.

```python
def loss_function(t, th):
    return sum(np.abs(t - th))
```

Now we want to test how well we can approximate our targets using a perceptron with a binary step function. 

Let us implement the binary step function first. This is given by

$$
\gamma_{\text{step}}(x) = \begin{cases}
1, & \text{if $x \geq 0$,} \\
0, & \text{otherwise.}
\end{cases}
$$

```python
def binary_step(x):
    return (x >= 0)
```

Now we have all we need to implement the perceptron. The perceptron is given by

$$
\begin{align}
\phi(x) &= \gamma_{\text{step}}(w_0 + \sum_{i=1}^2 w_i x_i) \\
        &= \gamma_{\text{step}}(w_0 + w^T x) \\
\end{align}
$$
i.e. the inner product or dot product of vectors $w, x$. 

Therefore, we need an intial set of weights $w$, which we initialize randomly from the normal distribution. Remember that the gradient descent optimization is parameterized by a learning rate $\eta$. We set our learning rate really small here, to avoid overstepping our solution. Also, we check if our solution is close enough to warrant an early stop in our iterations.

```python
wh_and = np.random.randn(3)
eta = 0.0001 # learning rate
scorehist = [] # History for plotting
epochs = 2000 # Number of epochs (iterations through entire data set)
for e in range(epochs):
    th = (X @ wh_and > 0) # Binary step function
    scorehist.append(loss_function(th, t_and))
    delta_w = eta * (th - t_and) @ X

    # Check early stopping
    if np.linalg.norm(delta_w, 2) < 1e-4:
        print("Target reached!")
        break
    
    wh_and -= delta_w

plt.plot(scorehist);
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss: AND");
```

Perfect. Our perceptron found a solution which is close enough to stop the number of iterations early, at approximately 1200 epochs.

We now want to visualize the solution space. Remember, this is a linear function. To represent the solution space, we create a meshgrid - a grid of 500 x 500 points in which we evaluate our perceptron and color the result accordingly.

```python
# Create meshgrid
_range = np.linspace(-1, 2, 500)
xs, ys = np.meshgrid(_range, _range)

# Define a function which evaluates the (non-activated) perceptron
def f(x, y, w):
    return w[0] + x * w[1] + y * w[2]
```

```python
# Draw contour plot
plt.contourf(xs, ys, f(xs, ys, wh_and), cmap="jet")
# Add color bar
plt.colorbar()

# Draw target points and classes
plt.scatter(data[t_and == 0, 0], data[t_and == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_and == 1, 0], data[t_and == 1, 1], color="red", edgecolors='black');
```

The yellow and red areas are classified as red dots (1s), and the green and blue areas are classified as cyan dots (0s). Notice that the decision area is linear, i.e. this is the slope of the perceptron without activation. Let us now visualize the activated outputs.

```python
# Define a function which evaluates the activated perceptron outputs
def f_active(x, y, w):
    return binary_step(w[0] + x * w[1] + y * w[2])

# Draw contour plot
plt.contourf(xs, ys, f_active(xs, ys, wh_and), cmap="jet", vmin=0, vmax=1)
# Add color bar
plt.colorbar()

# Draw target points and classes
plt.scatter(data[t_and == 0, 0], data[t_and == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_and == 1, 0], data[t_and == 1, 1], color="red", edgecolors='black');
```

This is the activated decision boundary. Looks pretty good! Most of the data points are correctly classified!

Now let us try the same with the XOR data. Remember, this is not linearly seperable, so it will be interesting to see how the perceptron draws a decision boundary.

```python
wh_xor = np.random.randn(3)
eta = 0.0001 # learning rate
scorehist = []
for e in range(epochs):
    th = (X @ wh_xor > 0)
    scorehist.append(loss_function(th, t_xor))
    delta_w = eta * (th - t_xor) @ X
    if np.linalg.norm(delta_w, 2) < 1e-4:
        print("Target reached!")
        break
    wh_xor -= delta_w

plt.plot(scorehist)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss: XOR");
```

No early stopping this time. Instead, the perceptron seems to wiggle around the weights a lot to rotate them in the best possible way, but ultimately not converging to a good approximation. Let's look at the results.

```python
# Draw contour plot
plt.contourf(xs, ys, f(xs, ys, wh_xor), cmap="jet")
# Add color bar
plt.colorbar()

# Draw target points and classes
plt.scatter(data[t_xor == 0, 0], data[t_xor == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_xor == 1, 0], data[t_xor == 1, 1], color="red", edgecolors='black');
```

This is hardly a good solution, wouldn't you agree? As the data has no linear seperator that makes sense, this is our best result.

Now let's try to swap the binary step function with the Logistic function, also known as the Sigmoid function. This maps the entire number line $\mathbb R$ to the interval $(0,1)$. The function is given by 

$$
\text{logistic}(x) = \sigma(x) =  \frac{1}{1 + e^{-x}}.
$$

The function is easily implemented as follows.

```python
def logistic(x):
    return 1/(1 + np.exp(-x))
```

Modifying the perceptron in this way makes it very similar to Logistic Regression. Let's try the perceptron again, this time with a higher learning rate.

```python
whl_and = np.random.randn(3)
eta = 0.1 # learning rate
scorehist = []
epochs = 2000
for e in range(epochs):
    th = logistic(X @ whl_and)
    scorehist.append(loss_function(th, t_and))
    delta_w = eta * (th - t_and) @ X
    if np.linalg.norm(delta_w, 2) < 1e-4:
        print("Target reached!")
        break
    whl_and -= delta_w

plt.plot(scorehist)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss: AND (Logistic Activation)");
```

Looks great, although the training loss drops off really quickly and spends a lot of time tweaking to find a solution which satisfies our stopping criteria. We could change the stopping criterion, but this is good enough for now. Let's look at the solution space.

```python
def f_logistic(x, y, w):
    return logistic(w[0] + x * w[1] + y * w[2])

plt.contourf(xs, ys, f_logistic(xs, ys, whl_and), cmap="jet", vmin=0, vmax=1)
plt.colorbar()
plt.scatter(data[t_and == 0, 0], data[t_and == 0, 1], color="cyan", edgecolors='black')
plt.scatter(data[t_and == 1, 0], data[t_and == 1, 1], color="red", edgecolors='black');
```

Lovely, isn't it? Notice that the logistic function is relatively shallow around the corners and steep at the decision boundary. Compared to the even, linear colors of the earlier plots without activation, we here see that the scaling is non-linear. This is why we call the logistic sigmoid function a non-linearity.
